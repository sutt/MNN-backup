```python
import xml
import xml.etree.ElementTree as ET
```

```python
root = ET.fromstring(s_xml)
```

```python
def parse_scoords(e):
    try:
        return [
            float(e.split(',')[0]), 
            float(e.split(',')[1])
        ]
    except:
        return None
```

```python
s_coords = root[0][0][0]

l_coords = s_coords.text.split('\n')

ll_coords = [parse_scoords(e) for e in l_coords]
ll_coords = [e for e in ll_coords if e is not None]
```
